```python
Datac = pd.DataFrame() # Dataframe de Resultados Totales
Metadatac = pd.DataFrame(columns=['Solicitud','ListaIDs','FechaEjecucion'\
                                  ,'Respuesta','ResultadosTotal','Resultadosxpagina']) # Dataframe de Informacion de Solicitudes

for id in ListaIDs[ithid:]: # For de conjuntos de 50 id's
  
  solicitud=solicitud+1
  
  if (solicitud == solicitudmax+1):
    ptoPartida = [] # Punto de partida para volver a comenzar el motor de busqueda en produccion.
    break

  time.sleep(espera) # Tiempo de espera por cada solicitud
  r = requests.get(URIv+key+'&part='+part+'&fields='+fields+'&id='+id) # Solicitud

  print('solicitud',solicitud,r.reason)

  if (r.reason == 'OK'): # Vemos si la solicitud fue exitosa
    
    d = dict(r.json())
    items = d.get('items')
    Data=pd.json_normalize(items).assign(Solicitud=solicitud)
    Datac=pd.concat([Datac, Data], ignore_index=True)  
    Metadatac = Metadatac.append({'Solicitud':solicitud,'ListaIDs':id,'FechaEjecucion':str(datetime.now())\
                                  ,'Respuesta':r.reason,'ResultadosTotal':d['pageInfo']['totalResults']\
                                  ,'Resultadosxpagina':d['pageInfo']['resultsPerPage']}, ignore_index=True)
  
  else: # Vemos si la solicitud no fue exitosa    
    print("No fue exitosa la solicitud", solicitud)

print("Videos Terminados")
```

```python
def cambiosEPM(texto):
  texto1=re.sub("elpatrondelmal", "el patron del mal", texto)
  texto2=re.sub("patrondelmal", "patron del mal", texto1)
  texto3=re.sub("patrondel", "patron del", texto2)
  texto4=re.sub("pabloescobar", "pablo escobar", texto3)
  texto5=re.sub("murtes", "muertes", texto4)
  texto6=re.sub("murte", "muerte", texto5)
  texto7=re.sub("muerts", "muertes", texto6)
  texto8=re.sub("patron", " patron ", texto7)
  texto9=re.sub("escobar", " escobar ", texto8)
  texto=re.sub("pablo", " pablo ", texto9) 
  return texto
```

##### Eliminando stopwords

```python
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
swsp=nltk.corpus.stopwords.words('spanish')
```

```python
def dropStopWordsSpanish(texto):
  if (str(texto) != 'nan'):
    tokens=word_tokenize(texto)
    tokensssw=[]
    for w in tokens:
      if w not in swsp:
        tokensssw.append(w)
    texto=" ".join(tokensssw)
  return texto
```

```python
swpo=nltk.corpus.stopwords.words('portuguese')
```

```python
def dropStopWordsPortugues(texto):
  if (str(texto) != 'nan'):
    tokens=word_tokenize(texto)
    tokensssw=[]
    for w in tokens:
      if w not in swsp:
        tokensssw.append(w)
    texto=" ".join(tokensssw)
  return texto
```

```python
swen=nltk.corpus.stopwords.words('english')
```

```python
def dropStopWordsEnglish(texto):
  if (str(texto) != 'nan'):
    tokens=word_tokenize(texto)
    tokensssw=[]
    for w in tokens:
      if w not in swen:
        tokensssw.append(w)
    texto=" ".join(tokensssw)
  return texto
```

```python
DATA['TituloVideoModificado']=DATA['TituloVideoModificado'].apply(dropStopWordsSpanish)
DATA['TituloVideoModificado']=DATA['TituloVideoModificado'].apply(dropStopWordsPortugues)
DATA['TituloVideoModificado']=DATA['TituloVideoModificado'].apply(dropStopWordsEnglish)
DATA['DescripcionVideoModificado']=DATA['DescripcionVideoModificado'].apply(dropStopWordsSpanish)
DATA['DescripcionVideoModificado']=DATA['DescripcionVideoModificado'].apply(dropStopWordsPortugues)
DATA['DescripcionVideoModificado']=DATA['DescripcionVideoModificado'].apply(dropStopWordsEnglish)
DATA['EtiquetasVideoTexto']=DATA['EtiquetasVideoTexto'].apply(dropStopWordsSpanish)
DATA['EtiquetasVideoTexto']=DATA['EtiquetasVideoTexto'].apply(dropStopWordsPortugues)
DATA['EtiquetasVideoTexto']=DATA['EtiquetasVideoTexto'].apply(dropStopWordsEnglish)
DATA['ConsultaVideoModificado']=DATA['ConsultaVideoModificado'].apply(dropStopWordsSpanish)
DATA['ConsultaVideoModificado']=DATA['ConsultaVideoModificado'].apply(dropStopWordsPortugues)
DATA['ConsultaVideoModificado']=DATA['ConsultaVideoModificado'].apply(dropStopWordsEnglish)
```
