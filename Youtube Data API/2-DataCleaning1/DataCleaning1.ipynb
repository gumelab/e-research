## Importamos datos en CSV (Para cada Serie)

```python
# ======================== OJO Modificar para cada serie ========================

path='/content/drive/MyDrive/AI-GUMELAB/CognASSISTANT-1/SEARCHES/YoutubeResults/Search/\
Narcos/2022-09-13SV(2022-09-12)CR(2022-09-13)/RESULTS-SEARCH-2022-09-13 13:37:07.521122.csv' 

# ======================== OJO Modificar para cada serie ========================
df = pd.read_csv(path)
df.info()
```

### Subconjunto de videos

```python
VD=df[df.idRecursoVD.notnull()].drop(['idRecursoPL', 'idRecursoCH'], axis=1)
VD.info()
```

### Subconjunto de PlayLists

```python
PL=df[df.idRecursoPL.notnull()].drop(['idRecursoVD', 'idRecursoCH'], axis=1)
PL.info()
```

### Subconjunto de Canales

```python
CH=df[df.idRecursoCH.notnull()].drop(['idRecursoVD', 'idRecursoPL'], axis=1)
CH.info()
```

## Eliminando resultados duplicados

### Videos

```python
VDclear=VD.drop_duplicates(subset=['idRecursoVD'])
VDclear.info()
```

### PlayLists

```python
PLclear=PL.drop_duplicates(subset=['idRecursoPL'])
PLclear.info()
```

### Canales

```python
CHclear=CH.drop_duplicates(subset=['idRecursoCH'])
CHclear.info()
```

## Exportando a CSV (Para cada Serie)

```python
ruta= '/content/drive/MyDrive/AI-GUMELAB/CognASSISTANT-1/SEARCHES/YoutubeResults/Search/'

# ======================== OJO Modificar para cada serie ========================

telenovela='Narcos/' 
fecha='2022-09-13SV(2022-09-12)CR(2022-09-13)/' 

# ======================== OJO Modificar para cada serie ========================
nombreVD='Videos'
nombrePL='PlayLists'
nombreCH='Canales'
VDclear.to_csv(ruta+telenovela+fecha+nombreVD+'.csv', index=False)
PLclear.to_csv(ruta+telenovela+fecha+nombrePL+'.csv', index=False)
CHclear.to_csv(ruta+telenovela+fecha+nombreCH+'.csv', index=False)
```
