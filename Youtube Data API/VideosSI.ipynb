### Importamos Videos

```python
# ======================== OJO Modificar para cada serie ========================

pathVDG='/content/drive/MyDrive/AI-GUMELAB/CognASSISTANT-1/SEARCHES/YoutubeResults/Series/\
TresCaines/Comentarios/RESULTS-PAISCOMENTARIOS(ComentariosGumelab)2022-05-23 19_36_24.415563.csv'

# ======================== OJO Modificar para cada serie ========================

dfVDG= pd.read_csv(pathVDG)
dfVDG.info()
```

```python
dfVDG=dfVDG[dfVDG.UtilComentarioPrincipalSi_No.notnull()]
```

```python
dfVDG['UtilComentarioPrincipalSi_No'] = dfVDG['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('mo', 'no'))
dfVDG['UtilComentarioPrincipalSi_No'] = dfVDG['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('so', 'si'))
dfVDG['UtilComentarioPrincipalSi_No'] = dfVDG['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('nsi', 'si'))
dfVDG['UtilComentarioPrincipalSi_No'] = dfVDG['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('mp', 'no'))
```

```python
# ======================== OJO Modificar para cada serie ========================

pathVD='/content/drive/MyDrive/AI-GUMELAB/CognASSISTANT-1/SEARCHES/YoutubeResults/Series/\
TresCaines/Comentarios/DataComentariosTotal.csv'

# ======================== OJO Modificar para cada serie ========================

dfVD= pd.read_csv(pathVD)
dfVD.info()
```

```python
dfVD=dfVD[dfVD.UtilComentarioPrincipalSi_No.notnull()]
```

```python
dfVD['UtilComentarioPrincipalSi_No'] = dfVD['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('mo', 'no'))
dfVD['UtilComentarioPrincipalSi_No'] = dfVD['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('so', 'si'))
dfVD['UtilComentarioPrincipalSi_No'] = dfVD['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('nsi', 'si'))
dfVD['UtilComentarioPrincipalSi_No'] = dfVD['UtilComentarioPrincipalSi_No'].apply(str.replace, args=('mp', 'no'))
```

```python
print("Clasificacion de videos por equipo Gumelab antes de filtrar\nSerie Tres Caines")
dfVDG['UtilComentarioPrincipalSi_No'].value_counts(ascending=False)
```

```python
print("Clasificacion de videos por equipo Gumelab despues de filtrar\nSerie Tres Caines")
dfVD['UtilComentarioPrincipalSi_No'].value_counts().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
from seaborn import palettes
plt.figure(figsize=(10,7))
dfVDG['UtilComentarioPrincipalSi_No'].value_counts(ascending=True).plot(kind='bar', color='limegreen', alpha=0.5, label='Antes de filtrar videos')
dfVD['UtilComentarioPrincipalSi_No'].value_counts(ascending=True).plot(kind='bar', color='skyblue', alpha=0.5, label='Despues de filtrar videos')
plt.title('Clasificación de comentarios por equipo Gumelab \n Serie Tres Caines')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
