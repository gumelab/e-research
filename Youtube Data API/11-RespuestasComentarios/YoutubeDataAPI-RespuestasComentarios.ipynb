#### Ejecucion

```python
Datac = pd.DataFrame() # Dataframe de Resultados Totales

Metadatac = pd.DataFrame(columns=['Solicitud','idComentarioPrincipal','maxResultados','FechaEjecucion','Respuesta'\
                                  ,'prevPageToken','nextPageToken','Resultadosxpagina']) # Dataframe de Informacion de Solicitudes

Faltantes=pd.DataFrame(columns=['idComentarioPrincipal','Solicitud'])

for idcomentario in ListIdComentarios[ithid:]: # For de id's de comentarios

  if (solicitud == solicitudmax+1):
    break

  filtro=dataCM[dataCM['idComentarioPrincipal'] == idcomentario]
  ReplicasTotales=list(filtro['ReplicasTotalesComentarioPrincipal'])[0]
  
  paginasmax = math.ceil(ReplicasTotales/100) # Numero de paginas maxinas para los comentarios
  pageToken = '' # Pagina especifica que se debe devolver OJO DEBE DE IR EN ESTE LUGAR

  for pagina in range(1,paginasmax+1): # For de la paginacion

    solicitud = solicitud+1
    
    if (solicitud == solicitudmax+1):
      print('Comentario id: ',idcomentario)
      ptoPartida = [idcomentario, pagina] # Punto de partida para volver a comenzar el motor de busqueda en produccion.
      break

    time.sleep(espera) # Tiempo de espera por cada solicitud
    r = requests.get(URIcr+key+'&part='+part+'&fields='+fields+'&parentId='+idcomentario\
                     +'&maxResults='+maxResults+'&pageToken='+pageToken+'&textFormat='+textFormat) # Solicitud
    
    print('solicitud',solicitud,r.reason)

    if (r.reason=='OK'): # Vemos si la solicitud fue exitosa

      d = dict(r.json())
      claves=list(d.keys())
      items = d.get('items')
      Data=pd.json_normalize(items).assign(Solicitud=solicitud)
      Datac=pd.concat([Datac, Data], ignore_index=True)

      if ('nextPageToken' in claves): # Si hay siguiente pagina
        pageToken=d['nextPageToken'] # Definimos el proximo pagetoken
        if ('prevPageToken' in claves):
          prevPageToken=d['prevPageToken']
        else:
          prevPageToken=''

        Metadatac = Metadatac.append({'Solicitud':solicitud,'idComentarioPrincipal':idcomentario,'maxResultados':maxResults\
                                      ,'FechaEjecucion':str(datetime.now()),'Respuesta':r.reason\
                                      ,'prevPageToken':prevPageToken,'nextPageToken':d['nextPageToken']\
                                      ,'Resultadosxpagina':d['pageInfo']['resultsPerPage']}, ignore_index=True)
        print("Siguiente pagina")

      else: # Si no hay siguiente pagina
        nextPageToken=''  # Definimos el nextPageToken como vacio
        if 'prevPageToken' in claves:
          prevPageToken=d['prevPageToken']
        else:
          prevPageToken=''
        
        Metadatac = Metadatac.append({'Solicitud':solicitud,'idComentarioPrincipal':idcomentario,'maxResultados':maxResults\
                                      ,'FechaEjecucion':str(datetime.now()),'Respuesta':r.reason\
                                      ,'prevPageToken':prevPageToken,'nextPageToken':nextPageToken\
                                      ,'Resultadosxpagina':d['pageInfo']['resultsPerPage']}, ignore_index=True)
        print("No hay mas paginas")
        break

    else: # Vemos si la solicitud no fue exitosa
      Faltantes=Faltantes.append({'idComentarioPrincipal': idcomentario,'Solicitud':solicitud}, ignore_index=True)
      print("La solicitud no se pudo realizar para el comentario ", idcomentario)
      break

print("Terminados los comentarios")
```

##### Eliminando stopwords

```python
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
swsp=nltk.corpus.stopwords.words('spanish')
```

```python
swpo=nltk.corpus.stopwords.words('portuguese')
```

```python
swen=nltk.corpus.stopwords.words('english')
```

```python
def dropStopWordsSpanish(texto):
  if (str(texto) != 'nan'):
    tokens=word_tokenize(texto)
    tokensssw=[]
    for w in tokens:
      if w not in swpo:
        tokensssw.append(w)
    texto=" ".join(tokensssw)
  return texto
```

```python
def dropStopWordsPortugues(texto):
  if (str(texto) != 'nan'):
    tokens=word_tokenize(texto)
    tokensssw=[]
    for w in tokens:
      if w not in swsp:
        tokensssw.append(w)
    texto=" ".join(tokensssw)
  return texto
```

```python
def dropStopWordsEnglish(texto):
  if (str(texto) != 'nan'):
    tokens=word_tokenize(texto)
    tokensssw=[]
    for w in tokens:
      if w not in swen:
        tokensssw.append(w)
    texto=" ".join(tokensssw)
  return texto
```
