## Representacion de texto o vectorizacion

```python
def cossmimilaridad(T,C):
  similaridad=[]
  for i in range(len(T)):
    similaridad.append(dot(T[:][i], C[:][i])/(norm(T[:][i])*norm(C[:][i])))
  return similaridad
```

### Bolsa de palabras

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
vect.fit(Videos['TituloEtiquetasModificado'])
```

```python
print("Vocabulary size: {}".format(len(vect.vocabulary_)))
print("Vocabulary content:\n {}".format(vect.vocabulary_))
```

```python
bag_of_wordsTextoTotal = vect.transform(Videos['TituloEtiquetasModificado'])
print("bag_of_words: {}".format(repr(bag_of_wordsTextoTotal)))
```

```python
bag_of_wordsConsulta = vect.transform(Videos['ConsultaVideoModificado'])
print("bag_of_words: {}".format(repr(bag_of_wordsConsulta)))
```

```python
T=bag_of_wordsTextoTotal.toarray()
C=bag_of_wordsConsulta.toarray()
```

```python
similaridad=cossmimilaridad(T,C)
```

```python
len(similaridad)
```

```python
SC=pd.Series(similaridad, name='Cos(BolsaPalabras)')
SC
```

```python
Videospuntaje=pd.concat([Videos,SC],axis=1)
Videospuntaje
```
