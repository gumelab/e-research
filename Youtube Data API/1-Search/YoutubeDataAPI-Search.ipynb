#### Constantes

```python
solicitud = 0 # Inicio de las solicitudes

# ======================= OJO Cambiar cuando el parametro es por relevancia paginasmax=2 =======================

paginasmax = 1 # Numero de paginas maximas

# ======================= OJO Cambiar cuando el parametro es por relevancia paginasmax=2 =======================

solicitudmax = 100 # Numero de solicitudes maximas (el limite es 100, pues cada solicitud cuesta 100 unidades y el maximo son 10000 unidades diarias.)
espera = 0.5 # Valor de retardo para las solicitudes

# ======================= OJO Cambiar diariamente =======================

ithConsulta = 14 # i-esima Consulta
ithPais = 0 # i-esimo Pais

# ======================= OJO Cambiar diariamente =======================
```

#### Ejecucion

```python
Datac= pd.DataFrame() # Dataframe de Resultados Totales
Metadatac = pd.DataFrame(columns=['Solicitud', 'maxResultados', 'Consulta', 'Pais', 'Lenguaje', 'FechaEjecucion'\
                                  , 'Respuesta', 'prevPageToken', 'nextPageToken', 'ResultadosTotal', 'Resultadosxpagina']) # Dataframe de Informacion de Solicitudes

for consulta in consultas[ithConsulta:]: # For de consultas para la serie

  if (solicitud == solicitudmax+1):
    break

  for pais in paises[ithPais:]: # For del pais donde se busca

    if (solicitud == solicitudmax+1):
      break

    pageToken='' # Pagina especifica que se debe devolver => OJO DEBE DE IR EN ESTE LUGAR

    for pagina in range(1,paginasmax+1): # For de la paginacion

      solicitud = solicitud+1

      if (solicitud == solicitudmax+1):
        ptoPartida = [consulta,pais,pagina] # Punto de partida para volver a comenzar el motor de busqueda en produccion.
        break

      time.sleep(espera) # Tiempo de espera por cada solicitud
      r = requests.get(URIs+key+'&part='+part+'&fields='+fields+'&channelType='+channelType\
                       +'&maxResults='+maxResults+'&order='+order\
                       +'&pageToken='+pageToken+'&q='+consulta\
                       +'&regionCode='+pais+'&relevanceLanguage='\
                       +relevanceLanguage+'&safeSearch='+safeSearch) # Solicitud

      print('solicitud',solicitud,r.reason)

      if (r.reason == 'OK'): # Vemos si la solicitud fue exitosa
        d = dict(r.json())
        claves = list(d.keys())
        items = d.get('items')
        Data = pd.json_normalize(items).assign(query = consulta,regionCode = pais,solicitud = solicitud)
        Datac = pd.concat([Datac, Data], ignore_index = True)

        if ('nextPageToken' in claves): # Si hay siguiente pagina
          pageToken = d['nextPageToken'] # Definimos el proximo pagetoken
          if ('prevPageToken' in claves):
            prevPageToken = d['prevPageToken']
          else:
            prevPageToken = ''
          Metadatac = Metadatac.append({'Solicitud':solicitud,'maxResultados':maxResults,'Consulta':consulta\
                          ,'Pais':pais,'Lenguaje':relevanceLanguage,'FechaEjecucion':str(datetime.now())\
                          ,'Respuesta':r.reason,'nextPageToken':d['nextPageToken'],'prevPageToken':prevPageToken\
                          ,'ResultadosTotal':d['pageInfo']['totalResults'],'Resultadosxpagina':d['pageInfo']['resultsPerPage']}, ignore_index = True)
          print("Siguiente pagina")

        else: # Si no hay siguiente pagina
          nextPageToken ='' # Definimos el nextPageToken como vacio
          if ('prevPageToken' in claves):
            prevPageToken = d['prevPageToken']
          else:
            prevPageToken = ''
          Metadatac = Metadatac.append({'Solicitud':solicitud,'maxResultados':maxResults,'Consulta':consulta\
                          ,'Pais':pais,'Lenguaje':relevanceLanguage,'FechaEjecucion':str(datetime.now())\
                          ,'Respuesta':r.reason,'nextPageToken':nextPageToken,'prevPageToken':prevPageToken\
                          ,'ResultadosTotal':d['pageInfo']['totalResults'],'Resultadosxpagina':d['pageInfo']['resultsPerPage']}, ignore_index = True)
          print("No hay mas paginas")
          break
      else: # Vemos si la solicitud no fue exitosa
        print("No fue exitosa la solicitud", solicitud)
print("Busqueda Terminada")
```
