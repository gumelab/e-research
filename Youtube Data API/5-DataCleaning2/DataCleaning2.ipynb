## Extrayendo Columnas de importancia

```python
Videos=dfVD.loc[:,['idRecursoVD','Consulta','Pais']]
Videos.info()
```

```python
PlayListVideos=dfPL.loc[:,['idRecursoVD','Consulta','Pais']]
PlayListVideos.info()
```

## Concatenando datos

```python
# si no hay videos de playlist ni canales
#data=pd.concat([Videos])
#data.info()
```

```python
# si no hay videos de canales
data=pd.concat([Videos,PlayListVideos])
data.info()
```

## Eliminando resultados duplicados

```python
dataclear=data.drop_duplicates(subset=['idRecursoVD'])
dataclear.info()
```
