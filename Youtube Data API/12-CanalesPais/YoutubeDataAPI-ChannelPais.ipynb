#### Eliminando Autores repetidos

```python
CMclear=dataCM.drop_duplicates(subset=[autor]).astype(str)
CMclear.info()
```

#### Formato Lista de IDs de Videos

```python
iz=0
de=50
IdsAutor=list(CMclear[autor])
ent=math.ceil(len(IdsAutor)/50)
ListaIDs=[]
for i in range(0,ent):
  stringIDs=','.join(IdsAutor[iz:de])
  iz=iz+50
  de=de+50
  ListaIDs.append(stringIDs)
print(len(ListaIDs))
```

#### Parametros constantes

```python
URIc='https://www.googleapis.com/youtube/v3/channels?key='

# ======parametros requeridos=====

part='snippet' #parametro obligatorio
fields='pageInfo(*),nextPageToken(*),prevPageToken(*)\
,items(id,snippet(country))'
```

#### Parametros variables

```python
#=====filtros (especifique exactamente uno de los siguientes parámetros)=====

id=ListaIDs #lista de ids de autores
```

#### Carpetas Drive

```python
today = str(date.today())

# ===== OJO Cambiar para cada Serie =====

PATH_TO_SAVE_SEARCHES = '/content/drive/MyDrive/AI-GUMELAB/CognASSISTANT-1/SEARCHES/YoutubeResults/PaisesCanales/RespuestasComentarios/Narcos/'

# ===== OJO Cambiar para cada Serie =====

PATH_TO_DIR_0 = PATH_TO_SAVE_SEARCHES + today
if os.path.exists(PATH_TO_DIR_0) == False:
  os.mkdir(PATH_TO_DIR_0)
file_nameL = 'PAISCOMENTARIOS'+str(datetime.now())
```

#### Constantes

```python
solicitud=0
espera=0 #valor de retardo para las solicitudes
ithid=0 # i-esimo id
solicitudmax=3000 #numero de solicitudes maximas (el limite es 10000)
```

#### Ejecucion

```python
datac= pd.DataFrame() # Dataframe de Resultados Totales
df = pd.DataFrame(columns=['Solicitud','ListaIDs','FechaEjecucion'\
                           ,'Respuesta'])
for id in ListaIDs[ithid:]:
  solicitud=solicitud+1
  time.sleep(espera)
  r = requests.get(URIc+key+'&part='+part+'&fields='+fields+'&id='+id)  
  print('solicitud',solicitud,r.reason)
  d = dict(r.json())
  items = d.get('items')
  data=pd.json_normalize(items)
  datac=pd.concat([datac, data], ignore_index=True)  
  df = df.append({'Solicitud':solicitud,'ListaIDs':id,'FechaEjecucion':str(datetime.now())\
                  ,'Respuesta':r.reason}, ignore_index=True)

print("Paises Terminados")
```
