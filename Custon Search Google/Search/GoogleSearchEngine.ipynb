#### Carpetas Drive 

```python
today = str(date.today())

# ======================= OJO Cambiar PATH para cada Serie =======================

PATH_TO_SAVE_SEARCHES = '/content/drive/MyDrive/AI-GUMELAB/CognASSISTANT-1/SEARCHES/GoogleResults/SearchResults/EscobarPatrondelMal/'

# ======================= OJO Cambiar PATH para cada Serie =======================

PATH_TO_DIR_0 = PATH_TO_SAVE_SEARCHES + today
if os.path.exists(PATH_TO_DIR_0) == False:
  os.mkdir(PATH_TO_DIR_0)
file_nameL = 'SEARCHDATOS'+str(datetime.now())
```

#### Constantes

```python
# ======================= OJO Cambiar diariamente =======================

ith_Criterio=498 # (Criterio = contenido historico + valoracion general) donde va comenzar la busqueda 

# ======================= OJO Cambiar diariamente =======================

solicitud = 0 # Inicio de las solicitudes
solicitudmax = 15 # Numero de solicitudes maximas (el limite es 100 de manera gratuita o 10000 con costo diario) #valor para detener la ejecucion bien sea por Criterios o por solicitudes.
espera = 1 # Valor de retardo para las solicitudes
START = [1,10,20,30,40,50,60,70,80,90] # Parametro donde comienza la paginacion (El numero maximo es 100 y start + num no puede ser mayor de 100)
```

#### Ejecucion

```python
Datac = pd.DataFrame() # Dataframe de Resultados Totales
Metadatac = pd.DataFrame() # Dataframe de Metadatos de Resultados Totales
Nosolicitudes = pd.DataFrame(columns=['Solicitud','Telenovela','CriterioNumero','Criterio'\
                                      ,'Consulta','Pais','Año2009_2021','Paginacion','FechaEjecucion']) # Dataframe con datos correspondientes a solicitudes malas.

for Criterio in Criterios[ith_Criterio:]: # Lista de Criterios

  CriterioNumero = Criterios.index(Criterio) # Numero de Criterio
  
  if (solicitud == solicitudmax + 1): 
    break
  
  q = Telenovela + Criterio # Consulta o query

  for ith_pais in range(len(gl)): # Lista de paises
    
    if (solicitud == solicitudmax + 1):
      break
    
    for ith_fecha in sort[:]: # Lista de Fechas
      
      if (solicitud == solicitudmax + 1):
        break
      
      for index_pagina in START[:]: # Lista de indices iniciales de paginacion
        
        solicitud = solicitud + 1

        if (solicitud == solicitudmax + 1):
          ptoPartida = [CriterioNumero, Criterio, gl[ith_pais], ith_fecha, index_pagina] # Punto de partida para volver a comenzar el motor de busqueda en produccion.
          break

        request = URI+key+'&cx='+cx+'&cr='+cr[ith_pais]+'&exactTerms='+exactTerms+'&filter='+filter+'&fields='+fields\
        +'&gl='+gl[ith_pais]+'&hl='+hl+'&lr='+lr+'&num='+num+'&orTerms='+orTerms+'&q='+q\
        +'&siteSearch='+siteSearch+'&siteSearchFilter='+siteSearchFilter+'&sort='+ith_fecha+'&start='+str(index_pagina)
        
        time.sleep(espera) # Tiempo de espera por cada solicitud
        r = requests.get(request) # Solicitud
        print('solicitud',solicitud,r.reason) # Respuesta

        if (r.reason == 'OK'): # Vemos si la solicitud fue exitosa
          d = dict(r.json())
          items = d.get('items')
          Metadata1 = pd.json_normalize(d['queries']['request'])
          Metadata2 = pd.json_normalize(d['searchInformation'])
          Metadata = pd.concat([Metadata1,Metadata2],axis = 1).assign(Telenovela = Telenovela, Consulta = q, CriterioBusqueda = Criterio\
                                                                      , CriterioNumero = CriterioNumero, FechaBusqueda = today\
                                                                      , Solicitud = solicitud)
          
          if items!=None: # Vemos si la solicitud exitosa tiene resultados
            Data = pd.json_normalize(items).assign(Telenovela = Telenovela, Consulta = q, CriterioBusqueda = Criterio, Criterio = 'C' + str(CriterioNumero)\
                                                   , Pais = Pais[gl[ith_pais]], Año2009_2021 = fecha[ith_fecha], Solicitud = solicitud)
            
            # ==Normalizar JSON en columna pagemap.metatags==      
            columnas = list(Data.columns)
            infodatac = pd.DataFrame()     
            for i in range(len(Data)):
              if ('pagemap.metatags' in  columnas):
                listaData=Data['pagemap.metatags'][i]
                if (type(listaData)==list):
                  if ('article:published_time' in list(listaData[0].keys())):
                    infodata = pd.json_normalize(listaData)
                    infodatac=pd.concat([infodatac,infodata],axis=0)
                  else:
                    infodata = pd.json_normalize([{'Texto':'No hay pagemap.metatags.fecha'}])
                    infodatac=pd.concat([infodatac,infodata],axis=0)
                else:
                  infodata = pd.json_normalize([{'Texto':'No hay pagemap.metatags.lista'}])
                  infodatac=pd.concat([infodatac,infodata],axis=0)
              else:
                infodata = pd.json_normalize([{'Texto':'No hay pagemap.metatags'}])
                infodatac=pd.concat([infodatac,infodata],axis=0)
      
            if ('article:published_time' in list(infodatac.columns)):
              info=infodatac[['Texto','article:published_time']]
            else:
              info=infodatac[['Texto']]
            info = info.reset_index()          
            # ==Normalizar JSON en columna pagemap.metatags==
            
            Dataglobal = pd.concat([Data, info],  axis = 1)
            Datac = pd.concat([Datac, Dataglobal], ignore_index = True) # Dataframe de Resultados Totales
            Metadata = Metadata.assign(Resultados = 'Hay Resultados')
            Metadatac = pd.concat([Metadatac, Metadata], ignore_index = True) # Dataframe de Metadatos donde no hay resultados
          
          else: # Vemos si la solicitud exitosa no tiene resultados
            Metadata = Metadata.assign(Resultados = 'No hay Resultados')
            Metadatac = pd.concat([Metadatac, Metadata], ignore_index = True) # Dataframe de Metadatos de Resultados Totales
            print("No hay links para Solicitud ",solicitud)
            break
            
        else: # Vemos si la solicitud no fue exitosa
          Norequest = [solicitud,Telenovela,CriterioNumero,Criterio,q,gl[ith_pais],ith_fecha,index_pagina,today]
          Nosolicitudes = Nosolicitudes.append({'Solicitud':solicitud, 'Telenovela':Telenovela, 'CriterioNumero':CriterioNumero\
                                                , 'Criterio':Criterio, 'Consulta':q, 'Pais':gl[ith_pais], 'Año2009_2021':ith_fecha\
                                                , 'Paginacion':index_pagina, 'FechaEjecucion':today},ignore_index = True)
          print("No se pudo realizadar la solicitud ",solicitud)

print("Termino la busqueda")
```
