### Eliminar links de facebook, instagram y twitter

```python
filterF=~Data_no_repeat['Sitios Web/ Fuente'].str.contains('facebook?', regex=True)
filterI=~Data_no_repeat['Sitios Web/ Fuente'].str.contains('instagram?', regex=True)
filterT=~Data_no_repeat['Sitios Web/ Fuente'].str.contains('twitter?', regex=True)
filter=((filterI) & (filterT) & (filterF))
```

```python
filter.value_counts()
```

```python
Data_no_repeat_noFIT=Data_no_repeat[filter]
Data_no_repeat_noFIT
```

```python
Data='DATOS'
Data_no_repeat_noFIT.to_csv(Ruta+file_nameL+Data+'.csv', index=False)
```
