```python
import pandas as pd

# Substitua 'caminho_para_o_arquivo.csv' pelo caminho do seu arquivo CSV
df = pd.read_csv('Tweets cruzados.csv')
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network

# Criação do grafo
G = nx.Graph()

# Adiciona os nós e arestas ao grafo a partir do DataFrame
for idx, row in df.iterrows():
    G.add_node(row['author_name'], title=row['author_name'])
    G.add_node(row['telenovela'], title=row['telenovela'])
    # Adiciona uma aresta com peso baseado em total_comentariomult
    G.add_edge(row['author_name'], row['telenovela'], weight=row['total_comentariomult'])


# Ou visualização interativa usando pyvis
net = Network(notebook=True)
net.from_nx(G)
net.show("network2.html")
```

```python
from pyvis.network import Network

# Inicializar o grafo de rede com pyvis
net = Network(notebook=True)

# Adiciona os nós e arestas ao grafo a partir do DataFrame
for idx, row in df.iterrows():
    # Adicionar ou atualizar o autor
    net.add_node(row['author_name'], title=row['author_name'], color="#f0a30a")
    
    # Adicionar ou atualizar a telenovela com um tamanho maior e cor diferente
    net.add_node(row['telenovela'], title=row['telenovela'], color="#7D3C98", size=300)
    
    # Adiciona uma aresta com peso baseado em total_comentariomult
    net.add_edge(row['author_name'], row['telenovela'], value=row['total_comentariomult'])

# Definir algumas opções adicionais
net.set_options("""
var options = {
  "nodes": {
    "font": {
      "size": 12
    }
  },
  "edges": {
    "color": {
      "inherit": true
    },
    "smooth": false
  },
  "physics": {
    "barnesHut": {
      "gravitationalConstant": -80000,
      "centralGravity": 0.3,
      "springLength": 95
    },
    "minVelocity": 0.75
  }
}
""")

# Salvar e exibir o gráfico interativo
net.show("network.html")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Suponha que df é o seu DataFrame carregado apropriadamente

# Crie o grafo
G = nx.Graph()

# Adicione os nós e arestas ao grafo a partir do DataFrame
for index, row in df.iterrows():
    G.add_node(row['author_name'], label=row['author_name'])
    G.add_node(row['telenovela'], label=row['telenovela'])
    G.add_edge(row['author_name'], row['telenovela'], weight=row['total_comentariomult'])

# Crie uma figura e um eixo
fig, ax = plt.subplots(figsize=(120, 120))

# Calcule a posição dos nós de forma automática
pos = nx.spring_layout(G)

# Desenhe os nós, arestas e rótulos
nx.draw_networkx_nodes(G, pos, ax=ax, node_size=15)
nx.draw_networkx_edges(G, pos, ax=ax)
nx.draw_networkx_labels(G, pos, ax=ax, font_size=8)

# Remova os eixos
ax.set_axis_off()

# Exiba o gráfico
plt.show()
```

```python
from pyvis.network import Network

# Inicializar o gráfico de rede com pyvis
net = Network(notebook=True, height="750px", width="100%")

# Adiciona os nós e arestas ao grafo a partir do DataFrame
for idx, row in df.iterrows():
    # Adiciona o nó do autor se ainda não foi adicionado
    if not net.get_node(row['author_name']):
        net.add_node(row['author_name'], title=row['author_name'], color="#f0a30a", size=15)

    # Adiciona o nó da telenovela se ainda não foi adicionado
    if not net.get_node(row['telenovela']):
        net.add_node(row['telenovela'], title=row['telenovela'], color="#7D3C98", size=25)

    # Adiciona uma aresta com peso baseado em total_comentariomult
    # E o título incluirá o nome do autor
    net.add_edge(row['author_name'], row['telenovela'], value=row['total_comentariomult'],
                 title=f"Author: {row['author_name']}, Comments: {row['total_comentariomult']}")

# Ajusta opções de visualização para melhorar a aparência e a interatividade
net.set_options("""
var options = {
  "nodes": {
    "shape": "dot",
    "size": 20,
    "font": {
      "size": 12
    }
  },
  "edges": {
    "width": 2,
    "shadow": true,
    "smooth": {
      "type": "continuous"
    }
  },
  "physics": {
    "forceAtlas2Based": {
      "gravitationalConstant": -26,
      "centralGravity": 0.005,
      "springLength": 230,
      "springConstant": 0.18
    },
    "maxVelocity": 146,
    "solver": "forceAtlas2Based",
    "timestep": 0.35,
    "stabilization": { "iterations": 150 }
  },
  "interaction": {
    "tooltipDelay": 300,
    "hideEdgesOnDrag": false
  }
}
""")

# Salvar e exibir o gráfico interativo
net.show("network3.html")
```

```python
from pyvis.network import Network

# Inicializar o gráfico de rede com pyvis
net = Network(notebook=True, height="750px", width="100%")

# Adiciona os nós e arestas ao grafo a partir do DataFrame
# Primeiro, vamos garantir que todos os nós sejam adicionados
nós_telenovelas = set(df['telenovela'].unique())
nós_authors = set(df['author_name'].unique())

# Adicionando nós de telenovelas com um tamanho e cor diferenciados
for telenovela in nós_telenovelas:
    net.add_node(telenovela, label=telenovela, color="#7D3C98", size=25)

# Adicionando nós de autores com uma cor padrão
for author in nós_authors:
    net.add_node(author, label=author, color="#f0a30a", size=15)

# Agora adicionamos as arestas
for idx, row in df.iterrows():
    net.add_edge(row['author_name'], row['telenovela'], value=row['total_comentariomult'],
                 title=f"Author: {row['author_name']}\nComments: {row['total_comentariomult']}")

# ... (restante do código para configurar e exibir o gráfico)
net.set_options("""
var options = {
  "nodes": {
    "shape": "dot",
    "size": 20,
    "font": {
      "size": 12
    }
  },
  "edges": {
    "width": 2,
    "shadow": true,
    "smooth": {
      "type": "continuous"
    }
  },
  "physics": {
    "forceAtlas2Based": {
      "gravitationalConstant": -26,
      "centralGravity": 0.005,
      "springLength": 230,
      "springConstant": 0.18
    },
    "maxVelocity": 146,
    "solver": "forceAtlas2Based",
    "timestep": 0.35,
    "stabilization": { "iterations": 150 }
  },
  "interaction": {
    "tooltipDelay": 300,
    "hideEdgesOnDrag": false
  }
}
""")

# Salvar e exibir o gráfico interativo
net.show("network3.html")
```
